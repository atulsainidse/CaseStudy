Let's plot JUNE for each subdivision for 1901 and 2015

```python
JUN = data.groupby('YEAR')['JUN'].apply(list)
```

```python
JUN
```

```python
type(JUN)
```

```python
year = list(range(1901,2016))
```

```python
year
```

```python
sum_of_rain = [sum(row) for row in JUN]
```

```python
fig_size = [0,0]
fig_size[0] = 15
fig_size[1] = 9
plt.rcParams["figure.figsize"] = fig_size
```

```python
plt.plot( year, sum_of_rain)
```

Let's explore Maharashtra......

Regions in maharashtra are - 'KONKAN & GOA', 'MADHYA MAHARASHTRA', 'MATATHWADA', 'VIDARBHA'

```python
MH_regions = ['KONKAN & GOA','MADHYA MAHARASHTRA', 'MATATHWADA', 'VIDARBHA']
```

```python
MH_data = data.loc[data.SUBDIVISION.isin(MH_regions)]
```

```python
len(MH_data) == 115 * 4
```

```python
MH_avg = MH_data.groupby(['SUBDIVISION']).agg(['mean'])
```

```python
MH_avg
```

Let's explore how much rain in Madhya Maharashtra over years

```python
M_MH = MH_data.loc[MH_data['SUBDIVISION'] == 'MADHYA MAHARASHTRA']
```

```python
M_MH.columns
```

```python
plt.plot( M_MH.YEAR, M_MH.JUN)
```
